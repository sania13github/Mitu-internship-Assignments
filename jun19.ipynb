```python
import matplotlib.pyplot as plt
```

```python
x = np.array ([23,34,42,49,55,61])
y = np.array ([25,36,47,52,56,65])
z = np.array ([87,99,78,94,76,69]) 
```

```python
#line plot
```

```python
plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker = 'o',mfc = 'green',color = 'black',ls = '--', ms = 7);
plt.grid()
```

```python
plt.title('Line Plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker = 'o',mfc = 'red',color = 'black',ls = '--', ms = 7);
plt.grid()
```

```python
courses = ['java','python','dart','R','C']
students = [67,63,59,54,47]
```

```python
plt.title('Bar plot')
plt.xlabel('courses')
plt.ylabel('students')
plt.bar(courses,students,color = 'skyblue',edgecolor = 'pink',width = 0.4,linewidth = 3);
plt.grid();
```

```python
plt.title('Bar plot')
plt.xlabel('courses')
plt.ylabel('students')
plt.bar(courses,students,color = 'skyblue',edgecolor = 'pink',width = 0.4,linewidth = 3);
plt.grid();
```
